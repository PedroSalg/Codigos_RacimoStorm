```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
Data = pd.read_csv('BodeIntegrator.csv', sep=',')
```

```python
Data. columns = ['F','Vin','Vout']
```

```python
Dat_log = 20*np.log10(Data['Vout'])
```

```python
plt.figure(figsize=(12,6))
plt.plot(Data['F'],Dat_log)
plt.xscale("log")
plt.grid(True, which="both",ls="-")
plt.xlabel("Frecuencia", fontsize=22)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
#plt.ylabel("Frecuencia", fontsize=22)
#plt.grid()
plt.savefig("Integrador_Bode.png")
```
