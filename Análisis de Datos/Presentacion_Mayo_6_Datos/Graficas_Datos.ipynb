```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import pandas as pd
import numpy as np
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes 
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
```

```python
Data = pd.read_csv('Prueba15.txt', sep=',')
Data. columns = ['Muestra']
plt.figure(figsize=(8,5))
plt.plot(Data['Muestra'],label='Prueba15.txt')
plt.grid()
plt.xlabel("Muestras", fontsize=22)
plt.ylabel("Nivel de Cuantizaci√≥n", fontsize=22)
plt.legend(fontsize=14)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
#plt.savefig()
```

```python
file1 = "Prueba11.txt"
Data1 = pd.read_csv(file1, sep=',')
Data1. columns = ['Muestra']

file2 = "Prueba3.txt"
Data2 = pd.read_csv(file2, sep=',')
Data2. columns = ['Muestra']

file3 = "Prueba4.txt"
Data3 = pd.read_csv(file3, sep=',')
Data3. columns = ['Muestra']

file4 = "Prueba5.txt"
Data4 = pd.read_csv(file4, sep=',')
Data4. columns = ['Muestra']
```

```python
fig1 = plt.figure(constrained_layout=True,figsize=(10.0,7.0))
gs = fig1.add_gridspec(6,6)

f1_ax1 = fig1.add_subplot(gs[0:3,0:3])
f1_ax1.plot(Data1['Muestra'], label=file1)
plt.legend()
plt.title("Trasmisi√≥n 1",fontsize=14)
plt.grid()
f1_ax1.set_ylabel("Nivel de cuantizaci√≥n",fontsize=10)
f1_ax1.set_xlabel("Muestras", fontsize=10)
f1_ax1.tick_params(labelsize='8' ) #labelcolor='r', width=3



f1_ax2 = fig1.add_subplot(gs[0:2,3:])
f1_ax2.plot(Data2['Muestra'], label=file2)
plt.legend()
plt.title("Trasmisi√≥n 2",fontsize=14)
plt.grid()
f1_ax2.set_ylabel("N. C.",fontsize=10)
f1_ax2.set_xlabel("Muestras", fontsize=10)
f1_ax2.tick_params(labelsize='8' ) #labelcolor='r', width=3

f1_ax3 = fig1.add_subplot(gs[2:4,3:])
f1_ax3.plot(Data4['Muestra'], label=file4)
plt.legend()
plt.title("Trasmisi√≥n 3",fontsize=14)
plt.grid()
f1_ax3.set_ylabel("N. C.",fontsize=10)
f1_ax3.set_xlabel("Muestras", fontsize=10)
f1_ax3.tick_params(labelsize='8' ) #labelcolor='r', width=3

f1_ax4 = fig1.add_subplot(gs[4:,:])
f1_ax4.plot(Data3['Muestra'], label=file3)
plt.legend()
plt.title("Trasmisi√≥n 4",fontsize=14)
plt.grid()
f1_ax4.set_ylabel("N. C.",fontsize=10)
f1_ax4.set_xlabel("Muestras", fontsize=10)
f1_ax4.tick_params(labelsize='8' ) #labelcolor='r', width=3

plt.savefig("Primeras_Trasmisiones.eps")
```

```python
Data = pd.read_csv('Prueba8.txt', sep=',')
Data. columns = ['Muestra']
fig_zoom = plt.figure(constrained_layout=True,figsize=(8.0,5.0))
#gs = fig1.add_gridspec(1,1)
ax = plt.axes()
ax.plot(Data['Muestra'],label='Prueba1.txt')
ax.grid()
ax.set_title("Verificaci√≥n Escritura", fontsize=24)
ax.set_ylabel("Nivel de Cuantizaci√≥n",fontsize=22)
ax.set_xlabel("Muestras", fontsize=22)
ax.tick_params(labelsize='16' ) #labelcolor='r', width=3
#ax.xlabel("Muestras", fontsize=22)
#plt.ylabel("Nivel de Cuantizaci√≥n", fontsize=22)
ax.legend(fontsize=14)
#plt.xticks(fontsize=16)
#plt.yticks(fontsize=16)
#plt.savefig()

# Zoom
x1 = 291.9
x2 = 298
# select y-range for zoomed region
y1 = 292
y2 = 300
# Make the zoom-in plot:
axins = zoomed_inset_axes(ax, 19, loc=5) # zoom = 2
axins.plot(Data['Muestra'])
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)
plt.xticks(visible=True)
plt.yticks(visible=True)
mark_inset(ax, axins, loc1=3, loc2=2, fc="none", ec="0.4")
#plt.draw()
#plt.show()

# Zoom
x3 = 271
x4 = 276
# select y-range for zoomed region
y3 = 272
y4 = 278

# Make the zoom-in plot:
axins = zoomed_inset_axes(ax, 25, loc=9) # zoom = 2
axins.plot(Data['Muestra'])
axins.set_xlim(x3, x4)
axins.set_ylim(y3, y4)
plt.xticks(visible=True)
plt.yticks(visible=True)
mark_inset(ax, axins, loc1=4, loc2=3, fc="none", ec="0.4")
#plt.draw()
#plt.show()

plt.savefig("Verficacion_Escritura_1.eps")
```
