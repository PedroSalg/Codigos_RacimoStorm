```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
Data = pd.read_csv('Prueba10.txt', sep=',')
Data. columns = ['Muestra']
```

```python
plt.figure(figsize=(8,5))
plt.plot(Data['Muestra'],label='Datos')
plt.grid()
plt.xlabel("Muestras", fontsize=22)
plt.ylabel("Nivel de Cuantización", fontsize=22)
plt.legend(fontsize=14)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
#plt.savefig()
```
